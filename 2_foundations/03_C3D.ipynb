```python
import torch

# Definimos una convolución 2D
# in_channels = 3 (por ejemplo RGB)
# out_channels = 2 (número de filtros)
# kernel_size = 3 (3x3)
m = torch.nn.Conv2d(3, 2, 3)

# Input:
# (batch, channels, height, width)
# batch = 1 (una imagen)
# channels = 3 (RGB)
# height = 5, width = 5
x = torch.randn(1, 3, 5, 5)

# Aplicamos la convolución
output = m(x)

# Output:
# (batch, out_channels, height, width)
# out_channels = número de filtros = 2
# height/width = (5 - 3 + 1) = 3  (sin padding, stride=1)
print("Output shape:", output.shape)

# Extra: ver forma de los pesos (kernels)
# (out_channels, in_channels, kernel_height, kernel_width)
print("Weight shape:", m.weight.shape)
```

```python
import torch

# Definimos una convolución 3D
# in_channels = 3
# out_channels = 2 (número de filtros)
# kernel_size = 3 (3x3x3)
m = torch.nn.Conv3d(3, 2, 3)

# Input:
# (batch, channels, depth, height, width)
# batch = 1
# channels = 3
# depth = 6 (por ejemplo: frames de video o slices)
# height = 5, width = 5
x = torch.randn(1, 3, 6, 5, 5)

# Aplicamos la convolución
output = m(x)

# Output:
# (batch, out_channels, depth, height, width)
# depth = (6 - 3 + 1) = 4
# height = (5 - 3 + 1) = 3
# width  = (5 - 3 + 1) = 3
print("Output shape:", output.shape)

# Extra: ver forma de los kernels
# (out_channels, in_channels, kernel_depth, kernel_height, kernel_width)
print("Weight shape:", m.weight.shape)
```
