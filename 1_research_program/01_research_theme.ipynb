# 🎓 Research Theme  

## Topic

Spatiotemporal Deep Learning for Video Surveillance Analysis

---

## Motivation

<div style="text-align: justify;">
Public safety remains a significant challenge in many countries, including Mexico. Automated systems capable of detecting abnormal or suspicious behaviors in real-time video could support surveillance systems and help reduce risks in public spaces. This research aims to contribute to the development of intelligent systems that assist in early detection of potentially dangerous situations.
</div>

---

## Proposed Approach

- Object detection (e.g., YOLO-based models)
- Multi-object tracking (e.g., ByteTrack or DeepSORT)
- Spatiotemporal modeling of video sequences using deep learning architectures (e.g., Transformers or 3D CNNs such as C3D)
- **Supervised action recognition for detecting known security-related events (e.g., fights and robberies)**
- **Extension toward anomaly detection to identify unseen or irregular behaviors in open-world scenarios**

---

## Datasets

- UCF-Crime (primary dataset for real-world surveillance scenarios)
- XD-Violence (complementary dataset for violence detection and robustness)
- ShanghaiTech (benchmark dataset for evaluation)
- Avenue (experimental dataset for validation)
- UCSD Ped2 (debugging and initial testing)

See [Datasets](../2_foundations/01_datasets_security.ipynb) for details.

---

## Expected Outcome

- A baseline system capable of detecting specific security-critical actions such as violence and robbery
- Extension of the system toward anomaly detection for identifying unseen abnormal behaviors
- Evaluation of model performance in terms of accuracy, generalization, and real-time capability

---

## Research Plan

- **Phase 0:** Study of foundational concepts and spatiotemporal models for video understanding, including key architectures such as 3D CNNs and Transformers, and review of essential literature in action recognition and anomaly detection

- **Phase 1:** Implementation of a baseline action recognition system for detecting simple events (e.g., fight vs normal) using short video clips

- **Phase 2:** Development of a more realistic surveillance pipeline by integrating object detection, multi-object tracking, and spatiotemporal modeling for detecting multiple actions (e.g., fight and robbery)

- **Phase 3:** Extension of the system toward anomaly detection to identify unseen or irregular behaviors in open-world surveillance scenarios

- **Phase 4:** Advanced Research and Model Development

- Development of hybrid models combining action recognition and anomaly detection
- Improvement of generalization across different environments and datasets
- Reduction of false positives in complex real-world scenarios
