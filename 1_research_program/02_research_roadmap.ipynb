# 🗺️ Research Roadmap

## Objective
Develop a real-time video surveillance system using spatiotemporal deep learning, progressing from action recognition to anomaly detection.

---

## Phase 0 — Foundations (1–2 months)

### Action Recognition (Fundamentals)
- *Two-Stream Convolutional Networks for Action Recognition in Videos* (Simonyan & Zisserman, 2014)
- *Learning Spatiotemporal Features with 3D Convolutional Networks (C3D)* (Tran et al., 2015)

### Modern Video Models (Transformers)
- *Is Space-Time Attention All You Need for Video Understanding? (TimeSformer)* (Bertasius et al., 2021)
- *Video Swin Transformer* (Liu et al., 2022)

### Anomaly Detection (Core Paper)
- *Real-World Anomaly Detection in Surveillance Videos* (Sultani et al., CVPR 2018)

### Goals
- Understand spatiotemporal representations
- Learn differences between action recognition and anomaly detection
- Become familiar with video datasets and evaluation metrics

---

## Phase 1 — Action Recognition (2–3 months)

### Datasets
- UCF-Crime

### Tasks
- Implement a baseline action recognition system:
  - fight vs normal classification
- Experiment with:
  - 3D CNNs (C3D)
  - Transformer-based models

### Goals
- Build a complete pipeline:
  - video → model → prediction
- Establish baseline performance

---

## Phase 2 — Realistic Surveillance System (3-4 months)

### Add Components
- Object detection:
  - YOLO-based models
- Multi-object tracking:
  - ByteTrack / DeepSORT

### Datasets
- UCF-Crime (primary)
- XD-Violence (complementary)

### Tasks
- Extend to multi-class detection:
  - fight
  - robbery
- Integrate spatial + temporal modeling

### Goals
- Build a surveillance-like system
- Improve robustness and performance

---

## Phase 3 — Anomaly Detection (4-6 months)

### Topics to Study
- Multiple Instance Learning (MIL)
- Autoencoders for anomaly detection
- Transformer-based anomaly detection

### Datasets
- UCF-Crime
- ShanghaiTech
- Avenue

### Tasks
- Implement anomaly detection models
- Generate anomaly scores over time

### Goals
- Detect unseen behaviors
- Handle weak supervision scenarios

---

## Phase 4: Advanced Research and Model Development (6–12+ months)

### Focus Areas
- Hybrid models (Action Recognition + Anomaly Detection)
- Real-time optimization (low latency, efficient models)
- Cross-domain generalization
- Reduction of false positives

### Goal
- Develop a novel contribution suitable for publication
