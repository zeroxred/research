# Two-Stream Convolutional Networks for Action Recognition in Videos (2014)

**Autores:** Karen Simonyan, Andrew Zisserman  
**Tema:** Reconocimiento de acciones en video usando Deep Learning

---

# Idea principal

El paper propone que el reconocimiento de acciones en video debe usar dos tipos de información:

1. **Información espacial**
   - Qué aparece en la imagen
   - Objetos, personas, escenas, poses

2. **Información temporal**
   - Cómo se mueve la imagen entre frames
   - Movimiento y dinámica de la acción

Para ello introducen una arquitectura de **dos streams**.

---

# Arquitectura Two-Stream

## 1. Spatial Stream

Entrada:
- Frames RGB normales

Objetivo:
- Aprender apariencia visual

La red funciona similar a una CNN clásica de clasificación de imágenes.

Aprende:
- personas
- objetos
- contexto
- poses

---

## 2. Temporal Stream

Entrada:
- Optical Flow entre frames consecutivos

Objetivo:
- Aprender movimiento

El optical flow representa:
- dirección del movimiento
- velocidad aparente
- patrones temporales

Esto ayuda a distinguir acciones similares visualmente.

Ejemplo:
- caminar
- correr
- lanzar

# Optical Flow

El paper usa optical flow denso como representación temporal.

Idea:
- calcular movimiento entre frames
- alimentar ese movimiento a una CNN

Ventaja:
- el modelo recibe información temporal explícita

Desventaja:
- calcular optical flow es costoso
- pipeline no completamente end-to-end

---

# Fusión de streams

Las predicciones de ambas redes se combinan al final.

Resultado:
- mejor desempeño que usar solo RGB
- mejor desempeño que usar solo movimiento

---

# Resultados

El método obtuvo resultados muy fuertes para la época en datasets como:

- UCF-101
- HMDB-51

Demostrando que:
- el movimiento es fundamental para reconocer acciones
- RGB por sí solo no es suficiente

---

# Aportaciones importantes

## 1. Introducción del enfoque Two-Stream
Separar:
- apariencia espacial
- información temporal

---

## 2. Uso de optical flow en Deep Learning para video
Popularizó usar optical flow como entrada de redes neuronales.

---

## 3. Influencia histórica
El paper influyó en:
- 3D CNNs
- SlowFast Networks
- modelos modernos de video
- Video Transformers

---

# Limitaciones

## Optical Flow costoso
El cálculo previo del flow consume mucho tiempo.

---

## Pipeline separado
El sistema no aprende completamente end-to-end.

---

## Escalabilidad limitada
Difícil entrenar eficientemente en datasets enormes.

---

# Impacto histórico

Es uno de los papers más influyentes en:
- action recognition
- video understanding
- deep learning aplicado a video

Introdujo una idea fundamental:
> En video importa tanto la apariencia como el movimiento.

---
