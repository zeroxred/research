# 🗺️ Research Roadmap

## Objective
Develop a real-time video surveillance system using spatiotemporal deep learning, progressing from action recognition to anomaly detection.

## Phase 0 — Foundations (3 months)

### Level 1 — Fundamentals

- **Learning Spatiotemporal Features with 3D Convolutional Networks**  
  Du Tran, Lubomir Bourdev, Rob Fergus, Lorenzo Torresani, Manohar Paluri (2015)

- **Two-Stream Convolutional Networks for Action Recognition in Videos**  
  Karen Simonyan, Andrew Zisserman (2014)

  **Note: [Conceptual/Theoretical]** These works will be studied to understand the transition from 2D to spatiotemporal modeling and the importance of temporal cues such as optical flow. They will serve as theoretical background rather than implementation targets.

### Level 2 — Advanced CNN-based Models

- **SlowFast Networks for Video Recognition**  
  Christoph Feichtenhofer, Haoqi Fan, Jitendra Malik, Kaiming He (2019)

  **Note: [Practical/Black-Box]** This model will be analyzed to understand multi-scale temporal processing. It will be used as a feature extractor with pre-trained weights, and its input-output representations will be examined for integration into the pipeline.

### Level 3 — Transformer-based Models

- **Is Space-Time Attention All You Need for Video Understanding?**  
  Gedas Bertasius, Heng Wang, Lorenzo Torresani (2021)

- **Video Swin Transformer**  
  Ze Liu, Jia Ning, Yue Cao, Yixuan Wei, Zheng Zhang, Stephen Lin, Baining Guo (2022)

  **Note: [Conceptual/Architectural]** The study will focus on the application of self-attention mechanisms to temporal modeling. Special attention will be given to computational complexity and its implications for real-time processing.

### Level 4 — Bridge to Anomaly Detection

- **Real-World Anomaly Detection in Surveillance Videos**  
  Waqas Sultani, Chen Chen, Mubarak Shah (2018)

  **Note: [Core/Preliminary Exploration]**  
  This work will be studied in detail to understand the formulation of anomaly detection in surveillance videos, including the Multiple Instance Learning (MIL) framework and dataset structure. Preliminary experiments may be conducted using a small subset of UCF-Crime to gain initial practical insight, without aiming for full reproduction at this stage. A full reproduction of the method is beyond the scope of this phase and may be addressed in a later stage of the research.

### Level 5 — State-of-the-Art & Efficiency (The 2026 Frontier)

- **Mamba: Linear-Time Sequence Modeling with Selective State Spaces**  
  Albert Gu, Tri Dao (2023)

- Recent works on State Space Models for video understanding (e.g., VideoMamba variants, 2024)

- Recent works on unified and multimodal video anomaly detection (2024–2025)

- Recent works on efficient temporal modeling for real-time video analysis (2024–2025)

  **Note: [Research Orientation]**  
  Recent approaches based on State Space Models (SSMs) and efficient architectures will be explored to identify emerging research directions. The focus will be on understanding their advantages and limitations, particularly in relation to real-time processing, rather than on implementation at this stage. A full reproduction of the method is beyond the scope of this phase and may be addressed in a later stage of the research.

### Goals
- Develop a solid understanding of spatiotemporal representations for video analysis
- Study and analyze key approaches in action recognition and anomaly detection to understand the current state of the art
- Identify the strengths and limitations of different modeling paradigms (e.g., CNN-based, Transformer-based, and emerging efficient architectures)
- Gain preliminary hands-on experience through limited and controlled experiments to support theoretical understanding
- Establish a clear research direction for subsequent phases, where more extensive implementations and experimentation may be conducted

## Phase 1 — Action Recognition (1-3 months)

### Datasets
- UCF-Crime

### Tasks
- Implement a baseline action recognition system:
  - Fight vs non-fight classification
- Experiment with:
  - 3D CNNs (C3D)
  - Transformer-based models

### Goals
- Build a complete pipeline:
  - video → model → prediction
- Establish baseline performance

---

## Phase 2 — Realistic Surveillance System (3-4 months)

### Add Components
- Object detection:
  - YOLO-based models
- Multi-object tracking:
  - ByteTrack / DeepSORT

### Datasets
- UCF-Crime (primary)
- XD-Violence (complementary)

### Tasks
- Extend to multi-class detection:
  - fight
  - robbery
- Integrate spatial + temporal modeling

### Goals
- Build a surveillance-like system
- Improve robustness and performance

---

## Phase 3 — Anomaly Detection (4-6 months)

### Topics to Study
- Multiple Instance Learning (MIL)
- Autoencoders for anomaly detection
- Transformer-based anomaly detection

### Datasets
- UCF-Crime
- ShanghaiTech
- Avenue

### Tasks
- Implement anomaly detection models
- Generate anomaly scores over time

### Goals
- Detect unseen behaviors
- Handle weak supervision scenarios

---

## Phase 4: Advanced Research and Model Development (6–12+ months)

### Focus Areas
- Hybrid models (Action Recognition + Anomaly Detection)
- Real-time optimization (low latency, efficient models)
- Cross-domain generalization
- Reduction of false positives

### Goal
- Develop a novel contribution suitable for publication
