# 🗂️ Video Anomaly & Action Detection Datasets

Dataset for security issues.

---

## **1. UCF-Crime**

[Official site](https://www.crcv.ucf.edu/projects/real-world/)

### **Description**
- ~1900 real-world surveillance videos (CCTV)
- 13 anomaly classes:
  - robbery
  - assault
  - burglary
  - vandalism
  - etc.

### **✅ Advantages**
- Real-world scenarios
- Includes key target events:
  - robbery
  - violence
- Widely used in SOTA research
- High relevance for security applications

### **❌ Limitations**
- Weak labels (no precise frame-level annotations)
- Long videos
- Class imbalance

### **Used in papers**
- Sultani et al., *Real-World Anomaly Detection in Surveillance Videos* (CVPR 2018)
- PoseWatch (2024)
- Video Anomaly Detection via LLMs (2024)
- Hierarchical Transformer VAD (2024)
- Global Temporal Transformer VAD (2025)
- Self-training Transformer VAD (2025)

### **Recommended usage**
👉 Primary dataset

## **2. XD-Violence**

[Official site](https://roc-ng.github.io/XD-Violence/)

### **Description**
- ~4000 videos
- Focused on:
  - violence
  - fights
- Multimodal (audio + video)

### **✅ Advantages**
- Large-scale dataset
- Strong for violence detection
- High diversity

### **❌ Limitations**
- Contains movie and YouTube content
- Less realistic than CCTV
- Potential domain bias

### **Used in papers**
- Video Anomaly Detection via LLMs (2024)
- Self-training Transformer for Weakly Supervised VAD (2025)
- Multiple multimodal anomaly detection works

### **Recommended usage**
👉 Complementary dataset

## 📊 Quick Comparison

| Dataset        | Realism | Violence | Robbery | Difficulty | Usage |
|---------------|--------|----------|---------|-----------|------|
| UCF-Crime     | ⭐⭐⭐⭐⭐ | ✔ | ✔ | High | 🥇 Primary |
| XD-Violence   | ⭐⭐⭐   | ✔ | ✖ | Medium | 🥈 Complementary |
| ShanghaiTech  | ⭐⭐⭐   | ✖ | ✖ | Medium | 🥉 Benchmark |
| Avenue        | ⭐⭐    | ✖ | ✖ | Low | Experimental |
| UCSD Ped2     | ⭐     | ✖ | ✖ | Very Low | Debug |

---

### **Column Explanation**

- **Realism**  
  Indicates how close the dataset is to real-world surveillance scenarios.  
  ⭐⭐⭐⭐⭐ = real CCTV footage, ⭐ = highly controlled environments.

- **Violence**  
  Whether the dataset includes violent actions such as fights, assaults, or physical aggression.

- **Robbery**  
  Whether the dataset includes theft or robbery-related events.

- **Difficulty**  
  Reflects how challenging the dataset is, considering:
  - annotation quality
  - video length
  - class imbalance  

- **Usage**  
  Recommended role:
  - Primary → main dataset  
  - Complementary → improves robustness  
  - Benchmark → comparison  
  - Experimental → quick tests  
  - Debug → pipeline validation

## **Key Insight**

- All datasets have limitations
- Research contribution lies in:
  - model architecture
  - temporal modeling
  - generalization capability

👉 Not in the dataset itself.
